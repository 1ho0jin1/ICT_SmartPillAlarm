```
import time
```

```
count = 0
for row in data.loc[0:100].iloc:
  print(row[["품목일련번호", "큰제품이미지", "표기이미지앞", "표기이미지뒤"]])
  wget.download(row["큰제품이미지"], out="./img_main/"+str(row["품목일련번호"])+"_main.jpg")
  if row["표기이미지앞"][0:4] == "http":
    wget.download(row["표기이미지앞"], out="./img_fptn/"+str(row["품목일련번호"])+"_fptn.gif")
  if row["표기이미지뒤"][0:4] == "http":
    wget.download(row["표기이미지뒤"], out="./img_bptn/"+str(row["품목일련번호"])+"_bptn.gif")
  count += 1
  time.sleep(1)
print(count)
```
